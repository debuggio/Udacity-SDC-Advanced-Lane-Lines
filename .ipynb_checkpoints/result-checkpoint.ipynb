## Advanced Lane Finding Project

The goals / steps of this project are the following:

* Compute the camera calibration matrix and distortion coefficients given a set of chessboard images.
* Apply a distortion correction to raw images.
* Use color transforms, gradients, etc., to create a thresholded binary image.
* Apply a perspective transform to rectify binary image ("birds-eye view").
* Detect lane pixels and fit to find the lane boundary.
* Determine the curvature of the lane and vehicle position with respect to center.
* Warp the detected lane boundaries back onto the original image.
* Output visual display of the lane boundaries and numerical estimation of lane curvature and vehicle position.

## First, I'll compute the camera calibration using chessboard images

```python
import numpy as np
import cv2
import glob
from moviepy.editor import VideoFileClip
import matplotlib.pyplot as plt
import re

%matplotlib qt
%matplotlib inline

from LineTracker import LineTracker
```

```python
# prepare object points, like (0,0,0), (1,0,0), (2,0,0) ....,(6,5,0)
def calibrate_camera():
    objp = np.zeros((6*9,3), np.float32)
    objp[:,:2] = np.mgrid[0:9,0:6].T.reshape(-1,2)

    # Arrays to store object points and image points from all the images.
    objpoints = [] # 3d points in real world space
    imgpoints = [] # 2d points in image plane.

    # Make a list of calibration images
    images = glob.glob('camera_cal/calibration*.jpg')
    img_shape = None
    
    # Step through the list and search for chessboard corners
    for idx, fname in enumerate(images):
        print("Processing " + fname + "...")
        img = cv2.imread(fname)
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

        img_shape = (img.shape[1], img.shape[0])
        
        # Find the chessboard corners
        ret, corners = cv2.findChessboardCorners(gray, (9,6),None)

        # If found, add object points, image points
        if ret == True:
            objpoints.append(objp)
            imgpoints.append(corners)

            # Draw and display the corners
            img = cv2.drawChessboardCorners(img, (9,6), corners, ret)
            cv2.imwrite("output_images\\corners_" + str(idx) + ".jpg", img)

    print(img_shape)
    ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, img_shape, None, None)

    cv2.destroyAllWindows()
    
    return mtx, dist
```

## Apply camera calibration

```python
camera_mtx, camera_dist = calibrate_camera()
```

```python
def undistort(img):
    return cv2.undistort(img, camera_mtx, camera_dist, None, camera_mtx)
```

### Thresholding

```python
def abs_sobel_thresh(img, orient='x', thresh_min=0, thresh_max=255, sobel_kernel=3):
    if orient not in ('x', 'y'):
        raise ValueError("orientation should be 'x' ot 'y'")
    # Apply the following steps to img
    # 1) Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # 2) Take the derivative in x or y given orient = 'x' or 'y'
    sobel = None
    if orient == 'x':
        sobel = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    else:
        sobel = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    # 3) Take the absolute value of the derivative or gradient
    abs_sobel = np.absolute(sobel)
    # 4) Scale to 8-bit (0 - 255) then convert to type = np.uint8
    scaled_sobel = np.uint8(255*abs_sobel/np.max(abs_sobel))
    # 5) Create a mask of 1's where the scaled gradient magnitude 
            # is > thresh_min and < thresh_max
    # 6) Return this mask as your binary_output image
    binary_output = np.zeros_like(scaled_sobel)
    binary_output[(scaled_sobel >= thresh_min) & (scaled_sobel <= thresh_max)] = 1
    #binary_output = np.copy(img) # Remove this line
    return binary_output
```

```python
def mag_thresh(img, sobel_kernel=11, mag_thresh=(0, 255)):
    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Take both Sobel x and y gradients
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    # Calculate the gradient magnitude
    gradmag = np.sqrt(sobelx**2 + sobely**2)
    # Rescale to 8 bit
    scale_factor = np.max(gradmag)/255 
    gradmag = (gradmag/scale_factor).astype(np.uint8) 
    # Create a binary image of ones where threshold is met, zeros otherwise
    binary_output = np.zeros_like(gradmag)
    binary_output[(gradmag >= mag_thresh[0]) & (gradmag <= mag_thresh[1])] = 1

    # Return the binary image
    return binary_output
```

```python
def dir_threshold(img, sobel_kernel=11, thresh=(0, np.pi/2)):
    
    # Apply the following steps to img
    # 1) Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # 2) Take the gradient in x and y separately
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    # 3) Take the absolute value of the x and y gradients
    # 4) Use np.arctan2(abs_sobely, abs_sobelx) to calculate the direction of the gradient 
    absarctan = np.arctan2(np.absolute(sobely), np.absolute(sobelx))
    # 5) Create a binary mask where direction thresholds are met
    binary_output =  np.zeros_like(absarctan)
    binary_output[(absarctan >= thresh[0]) & (absarctan <= thresh[1])] = 1
    # 6) Return this mask as your binary_output image
    #binary_output = np.copy(img) # Remove this line
    return binary_output
```

```python
def color_threshold(img, sthresh=(0,255), vthresh=(0,255)):
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    s = hls[:,:,2]
    s_binary_output = np.zeros_like(s)
    s_binary_output[(s >= sthresh[0]) & (s <= sthresh[1])] = 1
    
    hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
    v = hls[:,:,2]
    v_binary_output = np.zeros_like(v)
    v_binary_output[(v >= vthresh[0]) & (v <= vthresh[1])] = 1
    
    output = np.zeros_like(s)
    output[(s_binary_output == 1) & (v_binary_output == 1)] = 1
    
    return output
```

### Using threshholfing functions

```python
def process_image(image):
    img = undistort(image)
    
    processedImage = np.zeros_like(img[:,:,0])
    grad_x = abs_sobel_thresh(img, orient='x', thresh_min=25, thresh_max=255)
    grad_y = abs_sobel_thresh(img, orient='y', thresh_min=10, thresh_max=255)
    color_binary = color_threshold(img, sthresh=(100, 255), vthresh=(200, 255))
    processedImage[((grad_x == 1) & (grad_y == 1) | (color_binary == 1))] = 255

    return processedImage
```

```python
def warp_image(raw_image, processed_image):
    image_size = (raw_image.shape[1], raw_image.shape[0])
        
    bot_width = .76
    mid_width = .16
    height_pct = .66
    bottom_trim = .935
    offset = image_size[1]*.33

    src = np.float32([[raw_image.shape[1]*(.5-mid_width/2),raw_image.shape[0]*height_pct],[raw_image.shape[1]*(.5+mid_width/2),raw_image.shape[0]*height_pct],[raw_image.shape[1]*(.5+bot_width/2),raw_image.shape[0]*bottom_trim],[raw_image.shape[1]*(.5-bot_width/2),raw_image.shape[0]*bottom_trim]])
    dst = np.float32([[offset, 0], [image_size[0]-offset, 0], [image_size[0] - offset, image_size[1]], [offset, image_size[1]]])
    
    perspective_transform = cv2.getPerspectiveTransform(src, dst)
    inverse_perspective_transform = cv2.getPerspectiveTransform(dst, src)
    warped = cv2.warpPerspective(processed_image, perspective_transform, (image_size[0], image_size[1]), flags=cv2.INTER_LINEAR)
    
    return warped, perspective_transform, inverse_perspective_transform
```

```python
def find_window_centroids(warped, window_width=25, window_height=80, margin=25, smooth_factor=15):
    tracker = LineTracker(window_width=window_width, window_height=window_height, margin=margin, smooth_factor=smooth_factor)
    window_centroids = tracker.find_window_centroids(warped)
    
    return window_centroids
```

```python
def show_window_mask(width, height, image, center, level):
    result = np.zeros_like(image)
    result[int(image.shape[0]-(level+1)*height):int(image.shape[0]-level*height),max(0, int(center-width)):min(int(center+width), image.shape[1])] = 1
       
    return result
```

```python
def find_lane_lines(original_image, warped, inverse_perspective_transform, image_size):
    window_width = 25
    window_height = 120
    xm_per_pixel = 2/384
    ym_per_pixel = 5/720
    
    window_centroids = find_window_centroids(warped, window_width, window_height)
    
    l_points = np.zeros_like(warped)
    r_points = np.zeros_like(warped)
    
    leftx = []
    rightx = []
   
    #draw window for each level
    for level in range(0, len(window_centroids)):
        leftx.append(window_centroids[level][0])
        rightx.append(window_centroids[level][1])
        
        l_mask = show_window_mask(window_width, window_height, warped, window_centroids[level][0], level)
        r_mask = show_window_mask(window_width, window_height, warped, window_centroids[level][1], level)
        
        l_points[(l_points == 255) | ((l_mask == 1))] = 255
        r_points[(r_points == 255) | ((r_mask == 1))] = 255
        
    #draw results
    template = np.array(r_points+l_points, np.uint8)
    zero_channel = np.zeros_like(template)
    template = np.array(cv2.merge((zero_channel, template, zero_channel)), np.uint8)
    warpage = np.array(cv2.merge((warped, warped, warped)), np.uint8)
    result = cv2.addWeighted(warpage, 1, template, .75, .0)
    
    yvals = range(0, warped.shape[0])
    
    res_yvals = np.arange(warped.shape[0] - (window_height/2), 0, -window_height)
        
    left_fit = np.polyfit(res_yvals, leftx, 2)
    left_fitx = left_fit[0]*yvals*yvals + left_fit[1]*yvals + left_fit[2]
    left_fitx = np.array(left_fitx, np.int32)

    right_fit = np.polyfit(res_yvals, rightx, 2)
    right_fitx = right_fit[0]*yvals*yvals + right_fit[1]*yvals + right_fit[2]
    right_fitx = np.array(right_fitx, np.int32)    
    
    left_lane = np.array(list(zip(np.concatenate((left_fitx-window_width/2, left_fitx[::-1]+window_width/2), axis=0),np.concatenate((yvals,yvals[::-1]), axis=0))), np.int32)
    right_lane = np.array(list(zip(np.concatenate((right_fitx-window_width/2, right_fitx[::-1]+window_width/2), axis=0),np.concatenate((yvals,yvals[::-1]), axis=0))), np.int32)
    middle_marker = np.array(list(zip(np.concatenate((left_fitx+window_width/2, right_fitx[::-1]-window_width/2), axis=0),np.concatenate((yvals,yvals[::-1]), axis=0))), np.int32)
    
    road = np.zeros_like(img)
    road_bkg = np.zeros_like(img)
    
    left_line_color = [255, 0, 0]
    right_line_color = [0, 0, 255]
    middle_marker_color = [0, 255, 0]
    
    cv2.fillPoly(road, [left_lane], left_line_color)
    cv2.fillPoly(road, [right_lane], right_line_color)
    cv2.fillPoly(road, [middle_marker], middle_marker_color)
    
    road_warped = cv2.warpPerspective(road, inverse_perspective_transform, (image_size[0], image_size[1]), flags=cv2.INTER_LINEAR)

    road = cv2.addWeighted(original_image, 1, road_warped, .7, 0) 
       
    curve_fit_cr = np.polyfit(np.array(res_yvals, np.float32)*ym_per_pixel, np.array(leftx,np.float32)*xm_per_pixel, 2)
    curverad = ((1 + (2*curve_fit_cr[0]*res_yvals[1]*ym_per_pixel + curve_fit_cr[1])**2)**1.5) / np.absolute(2*curve_fit_cr[0])
    
    camera_center = (left_fitx[-1] + right_fitx[-1]) / 2

    center_diff = (camera_center-warped.shape[1]/2)*xm_per_pixel
    
    side_pos = 'left'
    if center_diff <= 0:
        side_pos = 'right'

    cv2.putText(road, 'Radius of Curvature = '+str(round(curverad,3))+'(m)',(50,50) , cv2.FONT_HERSHEY_SIMPLEX, 1,(5, 176, 249),2)
    cv2.putText(road, 'Vehicle is '+str(abs(round(center_diff,3)))+'m '+side_pos+' of center',(50,100) , cv2.FONT_HERSHEY_SIMPLEX, 1,(5, 176, 249),2)    
    
    return road
```

```python
def process_test_image(image, idx):
    image_size = (image.shape[1], image.shape[0])

    # undistort the image
    undistorted = undistort(image)
    write_name = './output_images/undistorted_'+str(idx)+'.jpg'
    cv2.imwrite(write_name, undistorted)

    # processed the image
    processed = process_image(image)
    write_name = './output_images/processed_'+str(idx)+'.jpg'
    cv2.imwrite(write_name, processed)

    # warped the image
    warped, perspective_transform, inverse_perspective_transform = warp_image(image, processed)
    write_name = './output_images/warped_'+str(idx)+'.jpg'
    cv2.imwrite(write_name, warped)    

    # warped the image
    tracked = find_lane_lines(image, warped, inverse_perspective_transform, image_size)
    write_name = './output_images/lines_'+str(idx)+'.jpg'
    cv2.imwrite(write_name, tracked)    
```

```python
def process_test_images():
    images = glob.glob('./test_images/test*.jpg')

    for idx, fname in enumerate(images):
        image = cv2.imread(fname)
        process_test_image(image, idx)
```

```python
def process_video_image(image):
    image_size = (image.shape[1], image.shape[0])
    
    undistorted = undistort(image)
    processed = process_image(image)
    warped, perspective_transform, inverse_perspective_transform = warp_image(image, processed)
    tracked = find_lane_lines(image, warped, inverse_perspective_transform, image_size)
    
    return tracked
```
